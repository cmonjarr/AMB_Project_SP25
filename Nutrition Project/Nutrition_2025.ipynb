#### **Addressing null/missing values**: 

**Determining best method for replacement**
The first step will be to identify how to best address any null values in the dataframes.

**Note**: For these datasets, the null value is in the "*Caffeine*" column of the *SBUX* dataframe. While the plan is to *delete* this column, below is an example of how to address null values for demonstrative purposes

**Method 1:  Converting null/missing values to *zero***:

*Note*: If the null/missing value represents unknown data or data that is truly missing, converting to zero may contribute to misleading results. It may be better to drop these rows/columns or use statistical methods.

Teas in the SBUX DataFrame were noted to have a non-numeric, variable ("Varies/varies") caffeine quantities and were also converted to zero.

```python
# one missing value was identified in the SBUX dataset in a column which will later be deleted.
# for this dataset type, null value could be changed to 0 as per below, but I would have personally replaced missing values with the mean of similar beverages using the following code:

#verify data type
print(sbux["Caffeine (mg)"].dtype)
#change "Varies/ varies" to a null value
sbux["Caffeine (mg)"] = sbux["Caffeine (mg)"].replace("Varies", 0)
sbux["Caffeine (mg)"] = sbux["Caffeine (mg)"].replace("varies", 0)
#convert null value to 0
sbux["Caffeine (mg)"] = sbux["Caffeine (mg)"].fillna(0)
#verify conversion / check for null values in this column
print(sbux["Caffeine (mg)"].isnull().sum())
```

**Method 2: Converting null/missing values value to *mean* (example of a statistical method)**:

For the scenario where a distinction between the "Iced Brewed Coffee (with Milk & Classic Syrup) with non-null values and the one with a null value was the type of milk (which does not contain caffeine), using an average of similar drinks could be a better replacement but does not account for variation in cup size. 

Again, this column will be deleted for the overall project.

*Note*: Data type and distribution are helpful in determining which statistical method to use.

For *numeric/quantitative data*, 
- **Mean**: best used when data is normally distributed (think bell curve) and there are no extrene outliers
- **Median**: best used when data either skews to the left or right or contains outliers
- **Mode**: helpful in show the most frequent occurences 

For *catgorical/qualitative data*,
- **Mode**: helpful in showing most common value

```python
#convert "Caffeine (mg)" column from object to integer
sbux["Caffeine (mg)"] = sbux["Caffeine (mg)"].astype(int)
#verify new column data type
print(sbux["Caffeine (mg)"].dtype)
```

```python
#determine mean "Caffeine (mg)" for Iced Brewed Coffee (With Milk & Classic Syrup)
similar_rows = sbux.loc["Iced Brewed Coffee (With Milk & Classic Syrup)"]
mean_caffeine = similar_rows['Caffeine (mg)'].mean()

print (mean_caffeine)
```

```python
# replace initial null value with mean "Caffeine (mg)" for similar beverages
sbux.loc[
    (sbux.index == "Iced Brewed Coffee (With Milk & Classic Syrup)") & (sbux["Beverage_prep"] == "2% Milk"),
    "Caffeine (mg)"
] = mean_caffeine
#verify change of initial null value
initial_null_row = sbux[
    (sbux.index == "Iced Brewed Coffee (With Milk & Classic Syrup)")  & (sbux["Beverage_prep"] == "2% Milk")
]
print(initial_null_row)
```

#### **Pandas Merge method**

 Merging two dataframes using outer join would follow the following code:
*combined_a = pd.merge(sbux,mcd, how='outer', left_index=True, right_index=True)*

Using the above method would ensure *all* rows are included in new dataframe. However, it works better for cominbining by matching keys or indices.

Since the two dataframes have been cleaned and prepared to contain similar columns, the concatenation method is a more effective approach to allow for comparison between the two restaurants.
