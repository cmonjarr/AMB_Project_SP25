**Addreessing missing values**: 

It is important to determine how to address any null values in the dataframes.

**Note**: For these datasets, the null value is in the "*Caffeine*" column of the *SBUX* dataframe. While the plan is to *delete* this column, below is an example of how to address null values.

One option is to convert null values to zero.

Teas were notable to have varianle caffeine intake and were also converted to zero.

```python
# one missing value was identified in the SBUX dataset in a column which will later be deleted.
# for this dataset type, null value could be changed to 0 as per below, but I would have personally replaced missing values with the mean of similar beverages using the following code:

#verify data type
print(sbux["Caffeine (mg)"].dtype)
#change "Varies/ varies" to a null value
sbux["Caffeine (mg)"] = sbux["Caffeine (mg)"].replace("Varies", 0)
sbux["Caffeine (mg)"] = sbux["Caffeine (mg)"].replace("varies", 0)
#convert null value to 0
sbux["Caffeine (mg)"] = sbux["Caffeine (mg)"].fillna(0)
#verify conversion / check for null values in this column
print(sbux["Caffeine (mg)"].isnull().sum())
```

In this scenario where a distinction between the "Iced Brewed Coffee (with Milk & Classic Syrup) with non-null values and the one with a null value was the type of milk (which does not contain caffeine), using an average of similar drinks could be a better replacement but does not account for variation in cup size. 

Below is an example of another approach to replacing the null value. Again, this column will be deleted for the overall project.

```python
#convert "Caffeine (mg)" column from object to integer
sbux["Caffeine (mg)"] = sbux["Caffeine (mg)"].astype(int)
#verify new column data type
print(sbux["Caffeine (mg)"].dtype)
```

### Merging/Joining: Combining two dataframes into one

**Pandas Merge method**


 Merging two dataframes using outer join would follow the following code:
*combined_a = pd.merge(sbux,mcd, how='outer', left_index=True, right_index=True)*

Using the above method would ensure *all* rows are included in new dataframe. However, it works better for cominbining by matching keys or indices.

Since the two dataframes have been cleaned and prepared to contain similar columns, the concatenation method is a more effective approach to allow for comparison between the two restaurants. 


**Concatenating method**

This method merges *all* rows from two dataframes that share the same structure (column names) aligning them along the 0 axis (i.e. index/"Beverage Info")

```python
#Concatenating two dataframes
combined = pd.concat([sbux, mcd], axis=0)
```

### Feature Engineering: Creating New Columns

**Calculating Percentages & Converting Ratios into New Columns**

```python
# creatng new columns for ratios of X-to-calories
combined['Chol_to_C'] = (combined['Cholesterol (mg)'] / combined['Calories'])
combined['Carb_to_C'] = (combined['Total Carbohydrates (g)'] / combined['Calories'])
combined['F_to_C'] = (combined['Dietary Fiber (g)'] / combined['Calories'])
combined['S_to_C'] = (combined['Sugars (g)'] / combined['Calories'])
combined['P_to_C'] = (combined['Protein (g)'] / combined['Calories'])
combined['A_to_C'] = (combined['Vitamin A (% DV)'] / combined['Calories'])
combined['C_to_C'] = (combined['Vitamin C (% DV)'] / combined['Calories'])
combined['Ca_to_C'] = (combined['Calcium (% DV)'] / combined['Calories'])
combined['Fe_to_C'] = (combined['Iron (% DV)'] / combined['Calories'])


# Data Preparation: converting new columns to percentages.  
combined['Chol_to_C'] = combined['Chol_to_C'].apply(lambda x: f"{x * 100:.2f}%")
combined['Carb_to_C'] = combined['Carb_to_C'].apply(lambda x: f"{x * 100:.2f}%")
combined['F_to_C'] = combined['F_to_C'].apply(lambda x: f"{x * 100:.2f}%")
combined['S_to_C'] = combined['S_to_C'].apply(lambda x: f"{x * 100:.2f}%")
combined['P_to_C'] = combined['P_to_C'].apply(lambda x: f"{x * 100:.2f}%")
combined['A_to_C'] = combined['A_to_C'].apply(lambda x: f"{x * 100:.2f}%")
combined['C_to_C'] = combined['C_to_C'].apply(lambda x: f"{x * 100:.2f}%")
combined['Ca_to_C'] = combined['Ca_to_C'].apply(lambda x: f"{x * 100:.2f}%")
combined['Fe_to_C'] = combined['Fe_to_C'].apply(lambda x: f"{x * 100:.2f}%")

# displaying updated dataframe
combined.head(5)
```
